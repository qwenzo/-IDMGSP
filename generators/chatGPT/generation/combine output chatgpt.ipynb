```python
import os
import re
files = os.listdir()
p = re.compile('output\d+_\d+.csv')
files = [ s for s in files if p.match(s) ]
files.sort()
files
```

```python
import pandas as pd
df_li = []

for filename in files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df_li.append(df)
```

```python
all_files_df = pd.concat(df_li, axis=0, ignore_index=True)[["title", "abstract", "introduction", "conclusion"]]
all_files_df
```

```python
all_files_df.to_csv('chatgpt_output_complete.csv') 
```
